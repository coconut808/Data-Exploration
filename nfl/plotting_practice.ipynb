```python
# read in the data
df = pd.read_csv("data/nfl_qb_data_2025_week4.csv")
df.head()
```

```python
print(df.info())
df.describe()
```

```python
df[['Player', 'Pass Yds', 'TD']].sort_values(by='Pass Yds', ascending=False).head(10)
```

```python
df.plot(
    kind='bar',
    title='QB Stats 2025 week 4',
    x='Player',
    y='Pass Yds',
    legend=True,
    fontsize=12
)
plt.show()
plt.savefig('images/qb_yards_week4.png', dpi=300)  # Save the figure with higher resolution
```

```python
df.plot(
    kind='bar',
    title='QB TD Stats 2025 week 4',
    x='Player',
    y='TD',
    legend=True,
    fontsize=12
)
plt.show()
plt.savefig('images/qb_td_week4.png', dpi=300)  # Save the figure with higher resolution
```
